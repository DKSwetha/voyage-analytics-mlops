```python
import pandas as pd
```

```python
df = pd.read_csv("../data/users.csv")
print(df.shape)
print(df.info())
```

```python
print(df["gender"].value_counts())
```

```python
labeled = df[df["gender"] != "none"]
unlabeled = df[df["gender"] == "none"]
print(f"Labeled rows: {len(labeled)}, Unlabeled rows: {len(unlabeled)}")
```

```python
print(labeled.groupby("gender")["age"].describe())
```

```python
print(pd.crosstab(labeled["company"], labeled["gender"]))
```
