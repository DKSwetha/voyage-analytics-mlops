```python
import pandas as pd
```

```python
users = pd.read_csv("../data/users.csv")
hotels = pd.read_csv("../data/hotels.csv")
flights = pd.read_csv("../data/flights.csv")
```

```python
print(hotels.groupby("place")["name"].unique())
# Every city has exactly ONE hotel - so "recommend hotel" = "recommend city"
```

```python
user_hotel_profile = hotels.groupby("userCode").agg(
    num_stays=("travelCode", "count"),
    avg_price_paid=("price", "mean"),
    avg_trip_length=("days", "mean"),
    favorite_place=("place", lambda x: x.mode()[0])
)
print(user_hotel_profile.describe())
```

```python
visits_per_user = flights.groupby("userCode")["to"].nunique()
print(visits_per_user.describe())
```
